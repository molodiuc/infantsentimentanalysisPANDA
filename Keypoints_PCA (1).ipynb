```python
import pandas as pd
import matplotlib.pyplot as plt
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# load dataset into Pandas DataFrame
df = pd.read_csv("C:\Users\RehabRoboticsLab\Desktop/all_keypoints3.csv")
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
features = ['1_x', '1_y','2_x', '2_y', '3_x', '3_y', '4_x', '4_y' ]
# Separating out the features
x = df.loc[:, features].values
# Separating out the target
y = df.loc[:,['target']].values
# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
x_values = df.loc[:,'normalized_x_keypoint'].values
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])
```

```python
x_values
```

```python
the_list = []

for k in range (0, 69):
    the_list.append([])
    the_list[0].append(x_values[k])
    
for k in range (69, 137):
    the_list.append([])
    the_list[1].append(x_values[k])
    
for k in range (137, 205):
    the_list.append([])
    the_list[2].append(x_values[k])

```

```python
import csv

csvfile = "C:\Users\RehabRoboticsLab\Desktop/all_keypoints3.csv"

#Assuming res is a flat list
with open(csvfile, "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for val in the_list:
        writer.writerow([the_list])
```

```python
finalDf = pd.concat([principalDf, df[['target']]], axis = 1)
finalDf.head()
finalDf.to_csv('PCA_Out1.csv')
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Positive', 'Negative', 'Neutral']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['target'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
